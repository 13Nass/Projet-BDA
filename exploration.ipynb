```python
# T1.0.1 ============= Chargement des CSV avec pandas =============
# importation des bibliothèques pour lire les csv et afficher des graphiques
import pandas as pd
import matplotlib.pyplot as plt

# création de variables qui contiennent les données de chaque csv
characters = pd.read_csv("imdb-small/characters.csv")
directors = pd.read_csv("imdb-small/directors.csv")
genres = pd.read_csv("imdb-small/genres.csv")
knownformovies = pd.read_csv("imdb-small/knownformovies.csv")
movies = pd.read_csv("imdb-small/movies.csv")
persons = pd.read_csv("imdb-small/persons.csv")
principals = pd.read_csv("imdb-small/principals.csv")
professions = pd.read_csv("imdb-small/professions.csv")
ratings = pd.read_csv("imdb-small/ratings.csv")
titles = pd.read_csv("imdb-small/titles.csv")
writers = pd.read_csv("imdb-small/writers.csv")
```

```python
#affichage d'une variable de données, les 5 premières données
print(characters.head())
```

```python
# j'ai créé une liste de données pour effectuer l'algo suivant :
files = {
    "characters": characters,
    "directors": directors,
    "genres": genres,
    "knownformovies": knownformovies,
    "movies": movies,
    "persons": persons,
    "principals": principals,
    "professions": professions,
    "ratings": ratings,
    "titles": titles,
    "writers": writers,
}

# à chaque données, on affiches sa taille, son type, le nombres de valeurs manquantes et le nombres de valeurs uniques
for name, df in files.items():
    print(f"\n\n===== {name} =====\n")
    print("— Nombre de lignes et colonnes\n", df.shape, "\n")
    print("— Types de données\n", df.dtypes, "\n")
    print(df.describe(include="all"), "\n")
    print("— Valeurs manquantes (NaN, chaînes vides)\n",
          df.isna().sum(), "\n",
          (df == "").sum(), "\n")
    print("— Valeurs uniques par colonne\n", df.nunique(), "\n")

```

```python
# Je change le nom des colonnes pour effectuer l'histogramme selon le nom des colonnes
movies.columns = ["mid", "titleType", "primaryTitle", "originalTitle",
                  "isAdult", "startYear", "endYear", "runtimeMinutes"]

print(movies.columns) 

# Histogramme du nombre de films par année
movies["startYear"] = pd.to_numeric(movies["startYear"], errors="coerce")
movies["startYear"].dropna().plot(kind="hist", bins=40)

movies["startYear"].dropna().plot(kind="hist", bins=40)
plt.xlabel("Année")
plt.ylabel("Nombre de films")
plt.title("Distribution des films par année")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Renommer les colonnes
movies.columns     = ["mid", "titleType", "primaryTitle", "originalTitle",
                      "isAdult", "startYear", "endYear", "runtimeMinutes"]
genres.columns     = ["mid", "genre"]
ratings.columns    = ["mid", "averageRating", "numVotes"]
principals.columns = ["mid", "ordering", "pid", "category", "job"]

print("=== Top 10 des genres les plus fréquents ===")

top_genres = genres["genre"].value_counts().head(10)
print(top_genres)

top_genres.plot(kind="bar")
plt.ylabel("Nombre de films")
plt.title("Top 10 des genres")
plt.show()
```

```python
print("=== Distribution des notes ===")

ratings["averageRating"].plot(kind="hist", bins=10)
plt.xlabel("Note")
plt.ylabel("Nombre de ratings")
plt.title("Distribution des notes")
plt.show()
```

```python
print("=== Nombre moyen d'acteurs par film ===")

actors_per_movie = principals.groupby("mid")["pid"].nunique()
mean_actors = actors_per_movie.mean()
print("Nombre moyen d'acteurs par film :", mean_actors)
```
